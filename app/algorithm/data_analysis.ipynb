```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import seaborn as sns
```

```python
courses = pd.read_csv('./course_cleaned.csv', header=0, encoding='unicode_escape')
```

```python
courses.head()
```

```python
courses.describe()
```

```python
courses.hist()
```

```python
sns.lmplot(data=courses, x='Term Yr',y='Enrolled',hue='Term',fit_reg=False)
```

```python
sns.histplot(data=courses, x='Class Yr',y='Cap', bins=10)
plt.xlabel('Enrolled')
plt.ylabel('Frequency')
plt.title('Histogram of Enrolled')
plt.show()
```

```python
# Filter the dataframe based on the condition
filtered_df = courses[courses['Term'] == 'Spring']

# Get distinct values of 'Course'
distinct_courses = filtered_df['Course'].unique()

# Print the distinct courses
print(distinct_courses)
```
