```python
import os
import pandas as pd
import numpy as np
```

```python
year1 = pd.read_csv(os.path.join(os.getcwd(), 'course_2019.csv'), header= 0,
                        encoding= 'unicode_escape')
year2 = pd.read_csv(os.path.join(os.getcwd(), 'course_2022.csv'), header= 0,
                        encoding= 'unicode_escape')
```

```python
year1.head()
```

```python
year2.head()
```

```python
year1.dropna(subset=['Faculty'], inplace=True)
year1 = year1[year1['Faculty'] == 'EN']
# Drop rows that have B or T or W in Section
year1 = year1[~year1['Section'].str.contains('B|T|W')]
# Drop Title from columns
year1.drop(['Title'], axis=1, inplace=True)
# Drop Sched Type from columns
year1.drop(['Sched Type'], axis=1, inplace=True)
# Drop Units from columns
year1.drop(['Units'], axis=1, inplace=True)
# Drop rows that have BME or CIVE or MECH in Subj
year1 = year1[~year1['Subj'].str.contains('BME|CIVE|MECH')]
year1.head()
```

```python
#Do same for year2 
year2.dropna(subset=['Faculty'], inplace=True)
year2 = year2[year2['Faculty'] == 'EN']
# Drop rows that have B or T or W in Section
year2 = year2[~year2['Section'].str.contains('B|T|W')]
# Drop Title from columns
year2.drop(['Title'], axis=1, inplace=True)
# Drop Sched Type from columns
year2.drop(['Sched Type'], axis=1, inplace=True)
# Drop Units from columns
year2.drop(['Units'], axis=1, inplace=True)
# Drop rows that have BME or CIVE or MECH in Subj
year2 = year2[~year2['Subj'].str.contains('BME|CIVE|MECH')]
year2.head()
```

```python
year1.to_csv(os.path.join(os.getcwd(), 'course_2019_cleaned.csv'), index=False)
year2.to_csv(os.path.join(os.getcwd(), 'course_2022_cleaned.csv'), index=False)
```
